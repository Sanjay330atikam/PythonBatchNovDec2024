# Shelve()
         Stores in .db files

```python
import shelve
```

```python
dir(shelve)
```

```python
# creating a new shelve file:
s= shelve.open("myshelve.db")# creates a python byte code file which is very hard to understand and edit
s['key1']={'int':10,'float':7.0,'string':'seven'}
s['person_details']=person_details = {
    'Name':'Roman Reigns',
    'Reach':69,
    'Weight':265,
    'Height':5.11,
    'Finisher':'Superman Punch'
    
}
s.close()
```

```python
# To access shelve file again open shelve file and use it like dictonary
s=shelve.open("myshelve.db")
existing =s['key1']
existing1=s['person_details']
print(existing)
print(person_details)
s.close()
```

```python
#. By default shevles do not track modifications to violate objects
with shelve.open('myshelve.db') as file:
    print("before:",file['key1'])
    print('before:',file['person_details'])

    file['key1']['new_value'] = 'This was not here before'
    # file['person_details']['new_value'] = 'This was not here before'
    print('after:',file['key1'])
    print('after:',file['person_details'])
```

```python
s =shelve.open('myshelve.db')# modifications are not reflected
print(s['key1'])
# print(s['person_details'])
s.close()
```

```python
# to automatically catch the changes open the shelve with writeback enabled
s=shelve.open('myshelve.db',writeback='True') 
print("before:",s['key1'])
print('before:',s['person_details'])
s['key1']['new_value'] = 'This was not here before'

print('after:',s['key1'])
print('after:',s['person_details'])
```
